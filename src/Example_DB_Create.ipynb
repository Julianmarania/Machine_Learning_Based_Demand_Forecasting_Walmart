## Librerias necesarias

```python
import pandas as pd
import os
import sqlite3

# pd.set_option('display.max_columns', None)
```

## Paso 1: Carga del conjunto de datos

```python
current_dir = os.getcwd()
calendar_path = os.path.join(current_dir, '../data/raw/calendar.csv')
sales_path = os.path.join(current_dir, '../data/raw/sales.csv')
sample_submission_path = os.path.join(current_dir, '../data/raw/sample_submission.csv')
sell_prices_path = os.path.join(current_dir, '../data/raw/sell_prices.csv')



calendar = pd.read_csv(calendar_path, sep = ',')
sales = pd.read_csv(sales_path, sep = ',')
sample_submission = pd.read_csv(sample_submission_path, sep = ',')
sell_prices = pd.read_csv(sell_prices_path, sep = ',')

calendar.head()
```

## Paso 2: Creación de una base de datos SQLite3

```python
#lo hago con with para asegurarme de que se cierre la conexión y no se bloquee. 

db_path = os.path.join(current_dir, '../data/data_base/DB_Walmart')

with sqlite3.connect(db_path) as con: # Creo, doy nombre y me conecto a la base de datos sqlite.
    calendar.to_sql('calendar', con, if_exists='replace', index=False)  # Inserto el dataframe en sql y lo combierto en una tabala de sql.
    sales.to_sql('sales', con, if_exists='replace', index=False) # Inserto el dataframe en sql y lo combierto en una tabala de sql.
    sample_submission.to_sql('sample_submission', con, if_exists='replace', index=False)# Inserto el dataframe en sql y lo combierto en una tabala de sql.
    sell_prices.to_sql('sell_prices', con, if_exists='replace', index=False)# Inserto el dataframe en sql y lo combierto en una tabala de sql.
```

```python
# Muestro los resultados para verificar que he insertado bien los datos en la base de datos sql.
with sqlite3.connect(db_path) as con:
    
    cursor = con.cursor() # Creo el cursor para ejecutar la consulta.
    
    cursor.execute('SELECT * FROM calendar') # Ejecuto la consulta SELECT para obtener todos los datos de la tabla.
     
    data_rows = cursor.fetchall() # Obtener todos los resultados.
    
    for row in data_rows: # Mostrar los resultados.
        print(row)
```

### Tras analizar las tablas decidimos que las tablas de valor y que extraeremos de la base de datos son: `calendar`, `sales`, y `sell_prices`.

## Paso 5: Realizamos una consulta y extraemos las columnas relevantes y las convertimos a dataframe

```python
calendar = data_calendar
sales = data_sales
sell_prices = data_sell_prices
```

#### Convertir `sales` a formato largo

```python
sales_long = sales.melt(
    id_vars = ["id", "item_id", "dept_id", "cat_id", "store_id", "state_id"],
    var_name = "d",
    value_name = "sales"
)
```

#### Unir `sales` con `calendar`

```python
sales_calendar = sales_long.merge(calendar, on = "d", how = "left")
```

#### Unir `sales_calendar` con `sell_prices`

```python
total_data = sales_calendar.merge(
    sell_prices,
    on = ["store_id", "item_id", "wm_yr_wk"],
    how = "left"
)
```

## Paso 6: Guardar dataframe como csv

```python
total_data_path = os.path.join(current_dir, '../data/raw/total_data.csv')
total_data.to_csv(total_data_path, index=False)
```
