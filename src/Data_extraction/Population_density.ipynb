## Population density

```python
import pandas as pd
import os
current_dir = os.getcwd()
```

```python
url_densidad_poblacion = os.path.join(current_dir, '../../data/raw/extractions/Densidad poblacion_y _Crecimiento_poblacional.xlsx')

# Abrir
df = pd.read_excel(url_densidad_poblacion, engine='openpyxl')

# Mostrar las primeras filas
print(df.head())
```

```python
print(df.columns.tolist)
```

```python
# Seleccionar columnas de estados y población 2018 (ajusta los nombres según tu archivo)
columnas_deseadas = ['table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)', 'Unnamed: 4', 'Unnamed: 5',
       'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9']  # Ejemplo: 'State' para estados y '2018' para datos de ese año
df_filtrado = df[columnas_deseadas]

# Eliminar filas que no son estados
df_filtrado = df_filtrado[~df_filtrado['table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)'].str.contains('United States|Puerto Rico', na=False)]
df_final = df_filtrado.iloc[8:].reset_index(drop=True)

# Cambiar nombres de columnas
df_densidad_poblacion = df_final.rename(columns={
    'table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)': 'State',     
    'Unnamed: 4' : 'poblacion_2011', 
    'Unnamed: 5' : 'poblacion_2012' ,
    'Unnamed: 6' : 'poblacion_2013',
    'Unnamed: 7' : 'poblacion_2014', 
    'Unnamed: 8' : 'poblacion_2015', 
    'Unnamed: 9' : 'poblacion_2016'
})

#Quitamos los . de los estados
df_densidad_poblacion['State'] = df_densidad_poblacion['State'].str.replace(r'^\.', '', regex=True)

# Mostrar resultado
df_densidad_poblacion.head()
```

```python
estados_deseados = ['California', 'Texas', 'Wisconsin']
df_densidad_filtrado = df_densidad_poblacion[df_densidad_poblacion['State'].isin(estados_deseados)].reset_index(drop=True)
df_densidad_filtrado
```

```python
# Diccionario de estados a abreviaturas
abreviaturas = {
    'California': 'CA',
    'Texas': 'TX',
    'Wisconsin': 'WI'
}

# Reemplazar los nombres por abreviaturas
df_densidad_filtrado['State'] = df_densidad_filtrado['State'].replace(abreviaturas)
df_densidad_filtrado
```

```python
## Guardar en CSV

df_densidad_filtrado_path = os.path.join(current_dir, '../../data/raw/macrodata/')

df_densidad_poblacion_data = os.path.join(df_densidad_filtrado_path, "population_density.csv")

df_densidad_filtrado.to_csv(df_densidad_poblacion_data, index=False)
```
