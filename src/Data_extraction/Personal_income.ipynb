```python
import os
import pandas as pd
```

```python
current_dir = os.getcwd()
personal_income_2011_path = os.path.join("../../data/raw/extractions/personal income 2011.csv")
personal_income_2012_path = os.path.join("../../data/raw/extractions/personal income 2012.csv")
personal_income_2013_path = os.path.join("../../data/raw/extractions/personal income 2013.csv")
personal_income_2014_path = os.path.join("../../data/raw/extractions/personal income 2014.csv")
personal_income_2015_path = os.path.join("../../data/raw/extractions/personal income 2015.csv")
personal_income_2016_path = os.path.join("../../data/raw/extractions/personal income 2016.csv")
```

```python
personal_income_2011 = pd.read_csv(personal_income_2011_path, sep=",", skiprows=3, nrows=3140)
personal_income_2012 = pd.read_csv(personal_income_2012_path, sep=",", skiprows=3, nrows=3140)
personal_income_2013 = pd.read_csv(personal_income_2013_path, sep=",", skiprows=3, nrows=3140)
personal_income_2014 = pd.read_csv(personal_income_2014_path, sep=",", skiprows=3, nrows=3140)
personal_income_2015 = pd.read_csv(personal_income_2015_path, sep=",", skiprows=3, nrows=3140)
personal_income_2016 = pd.read_csv(personal_income_2016_path, sep=",", skiprows=3, nrows=3140)
```

```python
from functools import reduce
dfs = [personal_income_2011, personal_income_2012, personal_income_2013, personal_income_2014, personal_income_2015, personal_income_2016]
df_merged = reduce(lambda left, right: pd.merge(left, right, on=['GeoFips', 'GeoName']), dfs)
```

```python
df_merged[df_merged['GeoName'].str.count(',') > 1]
# Divide por la última coma usando rsplit
df_merged[['City', 'State']] = df_merged['GeoName'].str.rsplit(', ', n=1, expand=True)
```

California, Texas, Wisconsin

```python
df_state = df_merged[df_merged["State"].isin(["WI","CA","TX"])]
df_state
```

```python
df_state[['2011', '2012', '2013', '2014', '2015', '2016']] = df_state[['2011', '2012', '2013', '2014', '2015', '2016']].apply(pd.to_numeric, errors='coerce')
```

```python
df_state
```

```python
df_state = df_state.drop(columns=["GeoFips"])
```

```python
df_state.isna().sum()
```

```python
df_state_yearly_mean = df_state.groupby("State")[['2011', '2012', '2013', '2014', '2015', '2016']].mean()
```

```python
# Añadir el prefijo "personal_income_mean_" a todas las columnas de año
df_state_yearly_mean.rename(columns=lambda col: f"pi_mean_{col}" if col.isdigit() else col, inplace=True)
df_state_yearly_mean = df_state_yearly_mean.reset_index()
df_state_yearly_mean
```

```python
## Guardar en CSV

df_state_yearly_mean_path = os.path.join(current_dir, '../../data/raw/macrodata/')

df_state_yearly_mean_data = os.path.join(df_state_yearly_mean_path, "personal_income.csv")

df_state_yearly_mean.to_csv(df_state_yearly_mean_data, index=False)
```
